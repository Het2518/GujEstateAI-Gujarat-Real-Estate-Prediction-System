## Cell 1 — Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.spines.top']   = False
plt.rcParams['axes.spines.right'] = False

print("Libraries imported successfully")
```

## Cell 2 — Load the Dataset

```python
df = pd.read_csv("../data/raw/ProjectInfo_Gujarat.csv")

print(f"Rows    : {df.shape[0]:,}")
print(f"Columns : {df.shape[1]}")
print()
df.head(3)
```

## Cell 3 — See All Column Names

```python
print("ALL COLUMNS:")
print("-" * 40)
for i, col in enumerate(df.columns, 1):
    print(f"  {i:>2}. {col}")
```

## Cell 4 — Data Types

```python
print("COLUMN DATA TYPES:")
print(df.dtypes.to_string())
```

## Cell 5 — Basic Statistics

```python
# min, max, mean, median for all number columns
df.describe().round(2)
```

## Cell 6 — Missing Values Analysis
Find which columns have empty/null data

```python
null_count   = df.isnull().sum()
null_percent = (null_count / len(df) * 100).round(2)

null_table = pd.DataFrame({
    "Missing Count"  : null_count,
    "Missing Percent": null_percent
})

null_table = null_table[null_table["Missing Count"] > 0]
null_table = null_table.sort_values("Missing Percent", ascending=False)

print("COLUMNS WITH MISSING VALUES:")
print("-" * 55)
for col, row in null_table.iterrows():
    pct = row["Missing Percent"]
    if pct > 40:
        tag = "DROP — too many nulls"
    elif pct > 10:
        tag = "CAREFUL — fill or drop"
    else:
        tag = "OK — small amount"
    print(f"  {col:<35} {pct:>6}%   {tag}")

clean = (df.isnull().sum() == 0).sum()
print(f"\n{clean} columns have ZERO missing values")
```

## Cell 7 — Project Type Distribution
How many projects exist per type?

```python
counts = df["projectType"].value_counts()

print("PROJECT TYPE COUNTS:")
print("-" * 45)
for ptype, count in counts.items():
    pct = count / len(df) * 100
    print(f"  {ptype:<30} {count:>5}  ({pct:.1f}%)")

# Bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(counts.index, counts.values,
               color=["#2196F3", "#4CAF50", "#FF9800", "#9C27B0"])

for bar in bars:
    h = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, h + 40,
             f"{int(h):,}", ha="center", fontweight="bold")

plt.title("Number of Projects by Type", fontsize=13, fontweight="bold")
plt.xlabel("Project Type")
plt.ylabel("Number of Projects")
plt.xticks(rotation=10)
plt.tight_layout()
plt.savefig("../reports/01_project_type.png", dpi=150)
plt.show()
```

## Cell 8 — Projects Per Year (Growth Trend)
How many projects registered each year?

```python
df_yr = df.dropna(subset=["startProjectYear"]).copy()
df_yr["startProjectYear"] = df_yr["startProjectYear"].astype(int)
df_yr = df_yr[(df_yr["startProjectYear"] >= 2010) & (df_yr["startProjectYear"] <= 2023)]

yearly = df_yr.groupby("startProjectYear").size().reset_index()
yearly.columns = ["Year", "Projects"]

print("PROJECTS REGISTERED EACH YEAR (2010-2023):")
print(yearly.to_string(index=False))

plt.figure(figsize=(11, 5))
plt.plot(yearly["Year"], yearly["Projects"],
         marker="o", linewidth=2.5, color="#2196F3",
         markersize=8, markerfacecolor="white", markeredgewidth=2.5)

for _, row in yearly.iterrows():
    plt.text(row["Year"], row["Projects"] + 25,
             str(int(row["Projects"])), ha="center", fontsize=8)

plt.title("Gujarat Real Estate — Projects Per Year", fontsize=13, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Number of Projects")
plt.xticks(yearly["Year"], rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.savefig("../reports/02_projects_per_year.png", dpi=150)
plt.show()
```

## Cell 9 — Top 15 Districts by Project Count

```python
dist_counts = df["distName"].value_counts().head(15)

print("TOP 15 DISTRICTS:")
print("-" * 40)
for dist, count in dist_counts.items():
    pct = count / len(df) * 100
    print(f"  {dist:<15} {count:>5}  ({pct:.1f}%)")

plt.figure(figsize=(9, 7))
plt.barh(dist_counts.index[::-1], dist_counts.values[::-1], color="#1565C0")

for i, val in enumerate(dist_counts.values[::-1]):
    plt.text(val + 20, i, f"{val:,}", va="center", fontsize=9)

plt.title("Top 15 Districts by Project Count", fontsize=13, fontweight="bold")
plt.xlabel("Number of Projects")
plt.tight_layout()
plt.savefig("../reports/03_top_districts.png", dpi=150)
plt.show()
```

## Cell 10 — Project Cost Distribution
How are project costs spread? (uses log scale for clarity)

```python
cost_cr = df["totalEstimatedCost"].dropna() / 1e7   # convert to Crores

print("PROJECT COST SUMMARY (Crores):")
print("-" * 38)
print(f"  Minimum  : Rs {cost_cr.min():>10,.2f} Cr")
print(f"  Median   : Rs {cost_cr.median():>10,.2f} Cr")
print(f"  Average  : Rs {cost_cr.mean():>10,.2f} Cr")
print(f"  Maximum  : Rs {cost_cr.max():>10,.2f} Cr")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(cost_cr, bins=60, color="#2196F3", edgecolor="white")
axes[0].set_title("Cost — Normal Scale (very skewed)")
axes[0].set_xlabel("Cost (Crores)")
axes[0].set_ylabel("Number of Projects")

axes[1].hist(np.log10(cost_cr + 1), bins=50, color="#4CAF50", edgecolor="white")
axes[1].set_title("Cost — Log Scale (clearer view)")
axes[1].set_xlabel("log10(Cost in Crores)")
axes[1].set_ylabel("Number of Projects")

plt.suptitle("Project Cost Distribution", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/04_cost_distribution.png", dpi=150)
plt.show()
```

## Cell 11 — Project Duration Analysis
How long do projects take? This will be our **Module 1 target variable**.

```python
df["duration_months"] = (
    (df["EndProjectYear"]  - df["startProjectYear"])  * 12 +
    (df["EndProjectMonth"] - df["startProjectMonth"])
)

df_d = df[(df["duration_months"] >= 3) & (df["duration_months"] <= 240)].copy()
d    = df_d["duration_months"]

print("PROJECT DURATION SUMMARY (months):")
print("-" * 40)
print(f"  Valid projects : {len(d):,}")
print(f"  Minimum        : {d.min():.0f} months")
print(f"  Median         : {d.median():.0f} months  ({d.median()/12:.1f} years)")
print(f"  Average        : {d.mean():.1f} months  ({d.mean()/12:.1f} years)")
print(f"  Maximum        : {d.max():.0f} months")

print("\nAVERAGE DURATION BY PROJECT TYPE:")
avg = df_d.groupby("projectType")["duration_months"].mean().round(1)
for pt, v in avg.items():
    print(f"  {pt:<30}  {v} months")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(d, bins=40, color="#FF9800", edgecolor="white")
axes[0].axvline(d.median(), color="red", linestyle="--",
                label=f"Median: {d.median():.0f} months")
axes[0].set_title("Project Duration Distribution")
axes[0].set_xlabel("Duration (months)")
axes[0].set_ylabel("Number of Projects")
axes[0].legend()

avg.plot(kind="bar", ax=axes[1],
         color=["#2196F3","#4CAF50","#FF9800","#9C27B0"])
axes[1].set_title("Avg Duration by Project Type")
axes[1].set_xlabel("Project Type")
axes[1].set_ylabel("Avg Duration (months)")
axes[1].tick_params(axis="x", rotation=12)

plt.suptitle("Project Duration Analysis", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/05_duration.png", dpi=150)
plt.show()
```

## Cell 12 — Booking Rate Analysis
What % of units are sold? This shows **market demand**.

```python
df["booking_rate"] = df["bookedUnits"] / df["totalUnits"]
b = df["booking_rate"].dropna()
b = b[(b >= 0) & (b <= 1)]

print("BOOKING RATE SUMMARY:")
print("-" * 38)
print(f"  Average rate     : {b.mean()*100:.1f}%")
print(f"  Median rate      : {b.median()*100:.1f}%")
print(f"  Fully booked     : {(b == 1.0).sum():,} projects (100% sold)")
print(f"  Low demand <20%  : {(b < 0.2).sum():,} projects")
print(f"  High demand >80% : {(b > 0.8).sum():,} projects")

print("\nAVG BOOKING RATE BY PROJECT TYPE:")
type_br = df.groupby("projectType")["booking_rate"].mean() * 100
for pt, v in type_br.items():
    print(f"  {pt:<30}  {v:.1f}%")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(b * 100, bins=30, color="#4CAF50", edgecolor="white")
axes[0].axvline(b.median()*100, color="red", linestyle="--",
                label=f"Median: {b.median()*100:.1f}%")
axes[0].set_title("Booking Rate Distribution")
axes[0].set_xlabel("Booking Rate (%)")
axes[0].set_ylabel("Number of Projects")
axes[0].legend()

type_br.plot(kind="bar", ax=axes[1],
             color=["#2196F3","#4CAF50","#FF9800","#9C27B0"])
axes[1].set_title("Avg Booking Rate by Project Type")
axes[1].set_xlabel("Project Type")
axes[1].set_ylabel("Booking Rate (%)")
axes[1].tick_params(axis="x", rotation=12)

plt.suptitle("Booking Rate Analysis", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/06_booking_rate.png", dpi=150)
plt.show()
```

## Cell 13 — Cost Per Sq.Ft and Outliers
Useful for detecting risky/anomalous projects in Module 5

```python
sqft     = df["avgCostPerSqFt"].dropna()
sqft     = sqft[sqft > 0]
p99      = sqft.quantile(0.99)
outliers = sqft[sqft > p99]

print("COST PER SQ.FT SUMMARY:")
print("-" * 38)
print(f"  Median    : Rs {sqft.median():>10,.0f}")
print(f"  Average   : Rs {sqft.mean():>10,.0f}")
print(f"  99th pct  : Rs {p99:>10,.0f}")
print(f"  Maximum   : Rs {sqft.max():>10,.0f}  (extreme outlier!)")
print(f"  Outliers above 99th% : {len(outliers)} projects")

sqft_clean = sqft[sqft <= p99]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(sqft_clean, bins=40, color="#9C27B0", edgecolor="white")
axes[0].set_title("Cost/Sq.Ft Distribution (extreme outliers removed)")
axes[0].set_xlabel("Rs per Sq.Ft")
axes[0].set_ylabel("Number of Projects")

df_box = df[df["avgCostPerSqFt"].between(1, p99)]
types  = df_box["projectType"].unique()
data   = [df_box[df_box["projectType"]==t]["avgCostPerSqFt"].values for t in types]
bp = axes[1].boxplot(data, patch_artist=True, labels=types)
for patch, c in zip(bp["boxes"], ["#2196F3","#4CAF50","#FF9800","#9C27B0"]):
    patch.set_facecolor(c)
axes[1].set_title("Cost/Sq.Ft by Project Type")
axes[1].set_ylabel("Rs per Sq.Ft")
axes[1].tick_params(axis="x", rotation=12)

plt.suptitle("Cost Per Sq.Ft Analysis", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/07_cost_sqft.png", dpi=150)
plt.show()
```

## Cell 14 — Correlation Heatmap
Which numeric columns are related to each other?

```python
cols = [
    "totalEstimatedCost",
    "totalLandCost",
    "totalDevelopCost",
    "totalSellingAmount",
    "totalUnits",
    "bookedUnits",
    "avgCostPerSqFt",
    "avgCostPerUnit",
    "totalCarpetArea_form3A",
]

corr = df[cols].dropna().corr().round(2)

print("STRONGEST CORRELATIONS (above 0.7):")
print("-" * 50)
for c1 in corr.columns:
    for c2 in corr.columns:
        if c1 < c2 and abs(corr.loc[c1, c2]) > 0.7:
            val = corr.loc[c1, c2]
            d   = "positive" if val > 0 else "negative"
            print(f"  {c1}  <->  {c2}  :  {val}  ({d})")

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="RdYlGn",
            center=0, linewidths=0.5, square=True)
plt.title("Correlation Heatmap", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/08_correlation_heatmap.png", dpi=150)
plt.show()
```

## Cell 15 — District Investment Overview

```python
df_d = df.dropna(subset=["distName","totalEstimatedCost"])

summary = df_d.groupby("distName").agg(
    Projects = ("projectRegId",       "count"),
    Total_Cr = ("totalEstimatedCost",  lambda x: round(x.sum() / 1e7, 1)),
).sort_values("Total_Cr", ascending=False)

print("TOP 15 DISTRICTS BY TOTAL INVESTMENT (Crores):")
print(summary.head(15).to_string())

top15 = summary.head(15)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].barh(top15.index[::-1], top15["Total_Cr"][::-1], color="#1565C0")
axes[0].set_title("Total Investment (Crores)")
axes[0].set_xlabel("Crores")

axes[1].barh(top15.index[::-1], top15["Projects"][::-1], color="#2E7D32")
axes[1].set_title("Number of Projects")
axes[1].set_xlabel("Projects")

plt.suptitle("District-Level Investment Overview", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.savefig("../reports/09_district_investment.png", dpi=150)
plt.show()
```

## Cell 16 — Final EDA Summary
Run this last — prints all key findings in one place.

```python
df["booking_rate"]    = df["bookedUnits"] / df["totalUnits"]
df["duration_months"] = (
    (df["EndProjectYear"]  - df["startProjectYear"])  * 12 +
    (df["EndProjectMonth"] - df["startProjectMonth"])
)
df_dur   = df[(df["duration_months"] >= 3) & (df["duration_months"] <= 240)]
b        = df["booking_rate"].dropna()
b        = b[(b >= 0) & (b <= 1)]
cost     = df["totalEstimatedCost"].dropna() / 1e7
null_pct = df.isnull().sum() / len(df) * 100
drop_cols = ", ".join(null_pct[null_pct > 40].index.tolist())

print()
print("=" * 58)
print("  GujEstateAI — EDA COMPLETE SUMMARY")
print("=" * 58)
print(f"  Total Projects : {len(df):,}")
print(f"  Total Columns  : {df.shape[1]}")
print(f"  Year Range     : {int(df['startProjectYear'].dropna().min())} to {int(df['startProjectYear'].dropna().max())}")
print()
print("  PROJECT TYPES")
for pt in df["projectType"].unique():
    n = (df["projectType"] == pt).sum()
    print(f"    {pt:<30}  {n:,}  ({n/len(df)*100:.1f}%)")
print()
print("  TOP 5 DISTRICTS")
for i, (dist, cnt) in enumerate(df["distName"].value_counts().head(5).items(), 1):
    print(f"    {i}. {dist:<15}  {cnt:,} projects")
print()
print("  COST (Crores)")
print(f"    Median : Rs {cost.median():.1f} Cr    Average : Rs {cost.mean():.1f} Cr")
print()
print("  DURATION")
print(f"    Average : {df_dur['duration_months'].mean():.1f} months ({df_dur['duration_months'].mean()/12:.1f} years)")
print(f"    Median  : {df_dur['duration_months'].median():.0f} months")
print()
print("  BOOKING RATE")
print(f"    Average : {b.mean()*100:.1f}%    Fully booked : {(b==1.0).sum():,} projects")
print()
print("  COLUMNS TO DROP (>40% nulls)")
print(f"    {drop_cols}")
print()
print("=" * 58)
print("  EDA Done. Ready for Phase 2: Data Cleaning")
print("=" * 58)
```
